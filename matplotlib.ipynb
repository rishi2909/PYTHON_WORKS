```python
#line plot
#whenever time component is involved on x-axis.
#depict the realtion b/w two varibles in the trend.

x = [1,2,3,4,5]
y = [5,1,7,9,2]
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.title("line plot to understand the syntax")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
# if i want to change the colour
plt.plot(x,y,color = "cyan")
plt.title("line plot to understand the syntax.")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
# if i want to change the colour
plt.plot(x,y,color = "cyan", marker = "o")
plt.title("line plot to understand the syntax.")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
# if i want to change the colour
plt.plot(x,y,color = "cyan", marker = "o", linestyle = '--')
plt.title("line plot to understand the syntax.")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
# if you want to increase the line width
plt.plot(x,y,color = "cyan", marker = "o", linestyle = '--', linewidth = 3)
plt.title("line plot to understand the syntax.")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
# if you want to inv marker size.
plt.plot(x,y,color = "cyan", marker = "o", linestyle = '--', linewidth = 3, markersize = 7)
plt.title("line plot to understand the syntax.")
plt.xlabel("x points")
plt.ylabel("y points")
plt.show()
```

```python
#if you wnat to add grid then grid >> horizonrtal,vertical line across graph
plt.plot(x,y,color = "cyan", marker = "o", linestyle = '--', linewidth = 3)
plt.title("line plot to understand the syntax.")
plt.xlabel("x points")
plt.ylabel("y points")
plt.grid()  
plt.show()
```

```python
# if you want to plot multiple lines on this plot
x = [1,2,3,4,5]
y1 = [10,5,7,9,4]
y2 = [2,9,5,1,10]
plt.plot(x,y1,color='r',marker='o',linestyle='--',linewidth=3, label = 'line y1')
plt.plot(x,y2,color='cyan',marker='o',linestyle='--',linewidth=3,label = 'liney2')
plt.xlabel("x data axis")
plt.ylabel("y data axis")
plt.title("Multiple Lines in one plot")
plt.legend() # it will help to see the label.
plt.show()
```

```python
# how to make a stacked histogram
data1= np.random.normal(0,1,5000)
data2 = np.random.normal(0,1,5000)
```

```python
data1
```

```python
data2
```

```python
plt.hist([data1,data2],bins=20,color=['cyan','yellow'],label=['dist1','dist2'])
plt.xlabel('Values')
plt.ylabel('frequency')
plt.title('stacked histogram')
plt.legend()
plt.show()
```

```python
# line chart >> to see the numerical features trend.
# pie chart >> help to understand the percentage share of particular coloumn.
langs = ['C++','Python','Java']
students = [60,70,50]

plt.pie(students,labels=langs)
plt.title('Pie Chart')
plt.show()
```

```python
# if you want to see % share then you 
langs = ['C++','Python','Java']
students = [60,70,50]

plt.pie(students,labels=langs, autopct = '%1.2f%%')
plt.title('Pie Chart')
plt.show()
```

```python
#if you want to break a pie chart then to see a particular pie
explode = (0.3,0.5,0.7)
langs = ['C++','Python','Java']
students = [60,70,50]

plt.pie(students,labels=langs, autopct = '%1.2f%%',explode = explode,shadow=True)
plt.title('Pie Chart')
plt.show()
```
